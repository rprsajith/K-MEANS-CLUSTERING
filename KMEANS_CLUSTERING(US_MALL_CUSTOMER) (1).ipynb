```python
x=df.iloc[:,3:5].values
x
```

```python
from sklearn.cluster import KMeans
em = []
for i in range(1,11):
  kmeans = KMeans(n_clusters=i, init="k-means++",max_iter=300,n_init=10,random_state=0)
  kmeans.fit(x)
  em.append(kmeans.inertia_)
plt.plot(range(1,11),em)
plt.title('elbow method')
plt.xlabel('clusters')
plt.ylabel('kmeans inertia')
plt.show()
```

```python
kmeans = KMeans(n_clusters=5,init="k-means++",random_state=0)
y_kmeans=kmeans.fit_predict(x)
```

```python
kmeans.inertia_
```

```python
centroid = kmeans.cluster_centers_
centroid
```

```python
labels=kmeans.labels_
print(labels)
```

```python
y_kmeans
```

```python
plt.figure(figsize=(15,7))
plt.scatter(x[y_kmeans == 0,0], x[y_kmeans == 0,1],color="green",label="Cluster 1",s=50)
plt.scatter(x[y_kmeans == 1,0], x[y_kmeans == 1,1],color="yellow",label="Cluster 2",s=50)
plt.scatter(x[y_kmeans == 2,0], x[y_kmeans == 2,1],color="red",label="Cluster 3",s=50)
plt.scatter(x[y_kmeans == 3,0], x[y_kmeans == 3,1],color="orange",label="Cluster 4",s=50)
plt.scatter(kmeans.cluster_centers_[:,0],kmeans.cluster_centers_[:,1],color='blue',label='Centroids',s=300,marker=',')
plt.grid(False)
plt.title('clusters of customers')
plt.xlabel('Annual Income(k$)')
plt.ylabel('Spending score(1-100)')
plt.legend()
plt.show()
```
